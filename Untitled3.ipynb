```python
# ==========================================================
# AI MENTAL HEALTH EMOTION INTELLIGENCE SYSTEM
# ==========================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

nltk.download('stopwords')
from nltk.corpus import stopwords

stop_words = set(stopwords.words("english"))

np.random.seed(42)

# ----------------------------------------------------------
# 1. SYNTHETIC EMOTION DATASET
# ----------------------------------------------------------

happy_samples = [
    "I feel amazing and motivated today",
    "Life is beautiful and I am grateful",
    "I am confident and excited about my future",
    "Today was productive and fulfilling",
    "I feel joyful and energized"
]

sad_samples = [
    "I feel lonely and exhausted",
    "Nothing seems to work in my life",
    "I am overwhelmed and stressed",
    "I feel hopeless and tired",
    "Today was depressing and draining"
]

anxious_samples = [
    "I am nervous about tomorrow",
    "I keep overthinking everything",
    "I feel restless and worried",
    "My heart feels heavy with stress",
    "I am anxious and tense"
]

data = []
for _ in range(200):
    data.append([np.random.choice(happy_samples), "Happy"])
    data.append([np.random.choice(sad_samples), "Sad"])
    data.append([np.random.choice(anxious_samples), "Anxious"])

df = pd.DataFrame(data, columns=["text", "emotion"])

print("Dataset Preview:")
print(df.head())

# ----------------------------------------------------------
# 2. TEXT CLEANING
# ----------------------------------------------------------

def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z ]', '', text)
    words = text.split()
    words = [w for w in words if w not in stop_words]
    return " ".join(words)

df["cleaned"] = df["text"].apply(clean_text)

# ----------------------------------------------------------
# 3. FEATURE EXTRACTION
# ----------------------------------------------------------

vectorizer = TfidfVectorizer(max_features=1000)
X = vectorizer.fit_transform(df["cleaned"]).toarray()
y = df["emotion"]

# ----------------------------------------------------------
# 4. TRAIN TEST SPLIT
# ----------------------------------------------------------

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# ----------------------------------------------------------
# 5. MODEL TRAINING
# ----------------------------------------------------------

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

# ----------------------------------------------------------
# 6. EVALUATION
# ----------------------------------------------------------

predictions = model.predict(X_test)

print("\nAccuracy:", accuracy_score(y_test, predictions))
print("\nClassification Report:\n")
print(classification_report(y_test, predictions))

cm = confusion_matrix(y_test, predictions)

plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, cmap="Purples", fmt='d')
plt.title("Emotion Classification Confusion Matrix")
plt.show()

# ----------------------------------------------------------
# 7. EMOTION DISTRIBUTION VISUALIZATION
# ----------------------------------------------------------

plt.figure(figsize=(6,5))
sns.countplot(x="emotion", data=df)
plt.title("Emotion Distribution")
plt.show()

# ----------------------------------------------------------
# 8. LIVE EMOTION ANALYZER
# ----------------------------------------------------------

print("\n------ LIVE EMOTION ANALYZER ------")
user_input = input("How are you feeling today? ")

cleaned_input = clean_text(user_input)
vector_input = vectorizer.transform([cleaned_input]).toarray()

prediction = model.predict(vector_input)[0]
confidence = max(model.predict_proba(vector_input)[0])

print("\nDetected Emotion:", prediction)
print("Confidence:", round(confidence * 100, 2), "%")

# ----------------------------------------------------------
# 9. MOOD TREND SIMULATION
# ----------------------------------------------------------

emotion_map = {"Happy": 1, "Sad": -1, "Anxious": -0.5}
df["emotion_score"] = df["emotion"].map(emotion_map)

plt.figure(figsize=(10,5))
plt.plot(df["emotion_score"].rolling(20).mean())
plt.title("Simulated Mood Trend Over Time")
plt.ylabel("Mood Score")
plt.show()
```
